```python
# Read in input data file
with open("input.txt","r") as f:
    geography_list = [x.strip() for x in f]
```

```python
# Inspect first several elements of geography
geography_list[0:2]
```

```python
# Function to count trees encountered
def trees_in_path(geography) :
    
    # Initialize count and position
    tree_count = 0
    position = 0
    
    len_pattern = len(geography[0])
    
    # Loop through each row
    for t in geography :
        if t[position] == "#" :
            tree_count += 1
        
        # increment over three for every row down
        position += 3
        
        # restart pattern if at end of line
        if position >= len_pattern :
            position -= (len_pattern)
        
    return tree_count
        
    
# Test on input file given
trees_in_path(geography_list)
```

```python
# Function to find correct number triple
def trees_in_path_slope(geography, slope) :
    
    # Pull in slope values
    num_right, num_down = slope[0], slope[1]
    
    # Initialize count and position
    tree_count = 0
    vertical = 0
    horizontal = 0
    
    len_pattern = len(geography[0])
    
    # Loop through each row
    for i, t in enumerate(geography) :
        if i != vertical :
            continue
            
        else :
            
            if t[horizontal] == "#" :
                tree_count += 1
        
            # increment over three for every row down
            horizontal += num_right
        
            # restart pattern if at end of line
            if horizontal >= len_pattern :
                horizontal -= (len_pattern)
                
            # move down specified number of rows
            vertical += num_down
        
    return tree_count
                
### Test on input file given

# First verify same results are produced for Part 1 slope
assert trees_in_path_slope(geography_list, (3, 1)) == trees_in_path(geography_list)

# Set up a list with slopes to calculate
slopes = [(1, 1), (3, 1), (5, 1), (7, 1), (1, 2)]

# Calculate number of trees for each slope
trees = [trees_in_path_slope(geography_list, s) for s in slopes]
print(trees)

# Multiply to get final answer
import math
math.prod(trees)
```
