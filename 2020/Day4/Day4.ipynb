```python
# Read in input data file
with open("input.txt","r") as f:
    passport_list = f.read().replace("\n"," ").split("  ") 
```

```python
# Inspect first several elements of geography
passport_list[0:2]
passport_list[-1] # note trailing space at end
```

```python
# Convert each line to dictionary
passport_dict_list = []

for p in passport_list :
    #print(p)
    p_dict = {}
    for x in p.strip().split(" ") :
        #print(x)
        #print(x.split(":"))
        key = x.split(":")[0]
        value = x.split(":")[1]
        p_dict[key] = value
    
    passport_dict_list.append(p_dict)

# Check on first few
passport_dict_list[0:3]
```

```python
# Function to count number of valid passports
def valid_id(passport) :
    valid_passports = []
    
    # Check all fields are present
    for p in passport :
        if ('byr' in p.keys() and 'iyr' in p.keys() and 'eyr' in p.keys() and
            'hgt' in p.keys() and 'hcl' in p.keys() and 'ecl' in p.keys() and
            'pid' in p.keys()) :
        
            valid_passports.append(True)
    
        else :
            valid_passports.append(False)
            
    return sum(valid_passports)
       
# Test on input file given
valid_id(passport_dict_list)
```

```python
# Function to count number of valid passports
def valid_id_values(passport) :
    import re
    
    valid_passports = []
    
    for p in passport :

        # First check presence of key fields
        if ('byr' in p.keys() and 'iyr' in p.keys() and 'eyr' in p.keys() and
            'hgt' in p.keys() and 'hcl' in p.keys() and 'ecl' in p.keys() and
            'pid' in p.keys()) :
            
            # Next check whether values are valid
            byr_valid = (int(p['byr']) >= 1920 and int(p['byr']) <= 2002)
            iyr_valid = (int(p['iyr']) >= 2010 and int(p['iyr']) <= 2020)
            eyr_valid = (int(p['eyr']) >= 2020 and int(p['eyr']) <= 2030)
            
            ecl_valid = (p['ecl'] in ['amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth'])
            
            hgt_pattern = re.compile('^\d{2,3}(in|cm)$')
            hgt_match = (hgt_pattern.match(p['hgt']) != None)
            if 'cm' in p['hgt'] :
                cm = p['hgt'][:-2]
                hgt_valid = (int(cm) >= 150 and int(cm) <= 193) and hgt_match == True
                
            elif 'in' in p['hgt'] :
                inch = p['hgt'][:-2]
                hgt_valid = (int(inch) >= 59 and int(inch) <= 76) and hgt_match == True
            
            else :
                hgt_valid = False
            
            hcl_pattern = re.compile('^#[a-f0-9]{6}$')
            hcl_valid = (hcl_pattern.match(p['hcl']) != None)
            
            pid_pattern = re.compile('^\d{9}$')
            pid_valid = (pid_pattern.match(p['pid']) != None)
            
            
            if (byr_valid == True and iyr_valid == True and eyr_valid == True and 
                ecl_valid == True and hgt_valid == True and hcl_valid == True and pid_valid == True) :
                
                valid_passports.append(True)
                
            else :
                valid_passports.append(False)
        
        # Key fields are not present
        else :
            valid_passports.append(False)
        
            
            
    return sum(valid_passports)

# Test on input file given
valid_id_values(passport_dict_list)
```
